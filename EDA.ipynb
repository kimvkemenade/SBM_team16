```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the data
path = "Crunchbase\Fintech\FinTechCompanies.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
# Printing all the columns to check which columns are needed
columns_list = df.columns.tolist()
for column in columns_list:
    print(column)
```

```python
# Filtering the columns based on the hypotheses
filtered_columns = [
    "Organization Name", "CB Rank (Organization)", "Industries", "Founded Date", "Year", "Founded Date Precision", "Estimated Revenue Range", 
    "BuiltWith - Active Tech Count", "G2 Stack - Total Products Active", # For H1
    "SimilarWeb - Monthly Visits Growth", "SimilarWeb - Visit Duration", # For H2
    "IPqwery - Patents Granted", "IPqwery - Trademarks Registered"] # For H3 

# Creating a new dataframe with only these columns
filtered_df = df[filtered_columns]
filtered_df
```

```python
# Checking the nan values for each column
nan_counts = filtered_df.isna().sum()
sorted_nan_counts = nan_counts.sort_values(ascending=False)
sorted_nan_counts_nonzero = sorted_nan_counts[sorted_nan_counts > 0]
for column, count in sorted_nan_counts_nonzero.items():
    print(f"{count}: {column}")
```

### Hypothesis 1: Companies that invest in differentiation (measured by higher active tech count, G2 Stack - Total Products Active) have a better CB rank

```python
# Creating a dataframe specifically for H1 where these columns are not nan
df_h1 = filtered_df.dropna(subset=["BuiltWith - Active Tech Count", "G2 Stack - Total Products Active"])
df_h1.head()
```

```python
df_h1["G2 Stack - Total Products Active"] = df_h1["G2 Stack - Total Products Active"].astype(int)
df_h1["BuiltWith - Active Tech Count"] = df_h1["BuiltWith - Active Tech Count"].astype(int)
if df_h1['CB Rank (Organization)'].dtype != int:
    df_h1['CB Rank (Organization)'] = df_h1['CB Rank (Organization)'].str.replace(',', '').astype(int)
```

```python
# Calculating the correlation for H1
correlation_active_tech_cb_rank = df_h1["CB Rank (Organization)"].corr(df_h1["BuiltWith - Active Tech Count"])
correlation_g2_stack_cb_rank = df_h1["CB Rank (Organization)"].corr(df_h1["G2 Stack - Total Products Active"])

print(f"Correlation between Active Tech Count and CB Rank: {correlation_active_tech_cb_rank}")
print(f"Correlation between G2 Stack and CB Rank: {correlation_g2_stack_cb_rank}")
```

```python
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('CB Rank (Organization)')
ax1.set_ylabel('Active Tech Count', color=color)
ax1.scatter(df_h1["CB Rank (Organization)"], df_h1["BuiltWith - Active Tech Count"], color=color, label='Active Tech Count')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Total Products Active', color=color)  # we already handled the x-label with ax1
ax2.scatter(df_h1["CB Rank (Organization)"], df_h1["G2 Stack - Total Products Active"], color=color, label='Total Products Active')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('CB Rank vs Active Tech Count and Total Products Active')
plt.show()
```

### Hypothesis 2: Companies that have higher monthly visit growth (SimilarWeb - Monthly Visits Growth, duration of time spent on platform) tend to have better CB Rank

```python
# Creating a dataframe specifically for H2 where these columns are not nan
df_h2 = filtered_df.dropna(subset=["SimilarWeb - Monthly Visits Growth", "SimilarWeb - Visit Duration"])
df_h2.head()
```

```python
# Converting the necessary columns to appropriate data types
if df_h2['SimilarWeb - Monthly Visits Growth'].dtype != float:
    df_h2['SimilarWeb - Monthly Visits Growth'] = df_h2['SimilarWeb - Monthly Visits Growth'].str.replace('%','').astype(float)
if df_h2["SimilarWeb - Visit Duration"].dtype != float:
    df_h2["SimilarWeb - Visit Duration"] = df_h2["SimilarWeb - Visit Duration"].str.replace(',', '').astype(float)
if df_h2['CB Rank (Organization)'].dtype != int:
    df_h2['CB Rank (Organization)'] = df_h2['CB Rank (Organization)'].str.replace(',', '').astype(int)
```

```python
# Calculating the correlations for H2
correlation_visits_growth_cb_rank = df_h2["CB Rank (Organization)"].corr(df_h2["SimilarWeb - Monthly Visits Growth"])
correlation_visit_duration_cb_rank = df_h2["CB Rank (Organization)"].corr(df_h2["SimilarWeb - Visit Duration"])

print(f"Correlation between Monthly Visits Growth and CB Rank: {correlation_visits_growth_cb_rank}")
print(f"Correlation between Visit Duration and CB Rank: {correlation_visit_duration_cb_rank}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot CB Rank vs Monthly Visit Growth
ax1.scatter(df_h2["CB Rank (Organization)"], df_h2["SimilarWeb - Monthly Visits Growth"], color='blue')
ax1.set_xlabel('CB Rank (Organization)')
ax1.set_ylabel('Monthly Visits Growth')
ax1.set_title('CB Rank vs Monthly Visits Growth')

# Plot CB Rank vs Visit Duration
ax2.scatter(df_h2["CB Rank (Organization)"], df_h2["SimilarWeb - Visit Duration"], color='green')
ax2.set_xlabel('CB Rank (Organization)')
ax2.set_ylabel('Visit Duration')
ax2.set_title('CB Rank vs Visit Duration')

plt.tight_layout()
plt.show()
```

### Hypothesis 3: Companies that invest more in IP (have more patents and trademarks) will have a higher CB rank 

```python
# Creating a dataframe specifically for H3 where these columns are not nan
df_h3 = filtered_df.dropna(subset=["CB Rank (Organization)", "IPqwery - Patents Granted", "IPqwery - Trademarks Registered"])
df_h3.head()
```

```python
# Convert the CB rank to an integer
if df_h3['CB Rank (Organization)'].dtype != int:
    df_h3['CB Rank (Organization)'] = df_h3['CB Rank (Organization)'].str.replace(',', '').astype(int)
df_h3['IPqwery - Patents Granted'] = df_h3['IPqwery - Patents Granted'].astype(str).str.replace(',', '')
df_h3['IPqwery - Trademarks Registered'] = df_h3['IPqwery - Trademarks Registered'].astype(str).str.replace(',', '')
df_h3['IPqwery - Patents Granted'] = pd.to_numeric(df_h3['IPqwery - Patents Granted'], errors='coerce')
df_h3['IPqwery - Trademarks Registered'] = pd.to_numeric(df_h3['IPqwery - Trademarks Registered'], errors='coerce')
```

```python
# Calculating the correlations for H3
correlation_patents_cb_rank = df_h3["CB Rank (Organization)"].corr(df_h3["IPqwery - Patents Granted"])
correlation_trademarks_cb_rank = df_h3["CB Rank (Organization)"].corr(df_h3["IPqwery - Trademarks Registered"])

print(f"Correlation between Patents Granted and CB Rank: {correlation_patents_cb_rank}")
print(f"Correlation between Trademarks Registered and CB Rank: {correlation_trademarks_cb_rank}")
```

```python
# Apply the cutoff
df_h3_cutoff = df_h3.copy()
df_h3_cutoff['IPqwery - Patents Granted'] = df_h3_cutoff['IPqwery - Patents Granted'].apply(lambda x: min(x, 100))
df_h3_cutoff['IPqwery - Trademarks Registered'] = df_h3_cutoff['IPqwery - Trademarks Registered'].apply(lambda x: min(x, 100))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot CB Rank vs Patents Granted
ax1.scatter(df_h3_cutoff["CB Rank (Organization)"], df_h3_cutoff["IPqwery - Patents Granted"], color='blue')
ax1.set_xlabel('CB Rank (Organization)')
ax1.set_ylabel('Patents Granted')
ax1.set_title('CB Rank vs Patents Granted')

# Plot CB Rank vs Trademarks Registered
ax2.scatter(df_h3_cutoff["CB Rank (Organization)"], df_h3_cutoff["IPqwery - Trademarks Registered"], color='green')
ax2.set_xlabel('CB Rank (Organization)')
ax2.set_ylabel('Trademarks Registered')
ax2.set_title('CB Rank vs Trademarks Registered')

plt.tight_layout()
plt.show()
```
