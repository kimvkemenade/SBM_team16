```python
import pandas as pd
```

```python
# Loading the data
path = "Crunchbase\Fintech\FinTechCompanies.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
# Printing all the columns to check which columns are needed
columns_list = df.columns.tolist()
for column in columns_list:
    print(column)
```

```python
# Filtering the columns based on the hypotheses
filtered_columns = [
    "Organization Name", "CB Rank (Organization)", "Industries", "Founded Date", "Year", "Founded Date Precision", "Estimated Revenue Range", 
    "BuiltWith - Active Tech Count", "G2 Stack - Total Products Active", # For H1
    "SimilarWeb - Monthly Visits Growth", "SimilarWeb - Visit Duration", # For H2
    "IPqwery - Patents Granted", "IPqwery - Trademarks Registered"] # For H3 

# Creating a new dataframe with only these columns
filtered_df = df[filtered_columns]
filtered_df
```

```python
# Checking the nan values for each column
nan_counts = filtered_df.isna().sum()
sorted_nan_counts = nan_counts.sort_values(ascending=False)
sorted_nan_counts_nonzero = sorted_nan_counts[sorted_nan_counts > 0]
for column, count in sorted_nan_counts_nonzero.items():
    print(f"{count}: {column}")
```

```python
# Creating a dataframe specifically for H1 where these columns are not nan
df_h1 = filtered_df.dropna(subset=["BuiltWith - Active Tech Count", "G2 Stack - Total Products Active"])
df_h1
```

```python
# Creating a dataframe specifically for H2 where these columns are not nan
df_h2 = filtered_df.dropna(subset=["SimilarWeb - Monthly Visits Growth", "SimilarWeb - Visit Duration"])
df_h2
```

```python
# Creating a dataframe specifically for H3 where these columns are not nan
df_h3 = filtered_df.dropna(subset=["IPqwery - Patents Granted", "IPqwery - Trademarks Registered"])
df_h3
```
